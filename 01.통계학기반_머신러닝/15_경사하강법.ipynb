# 경사하강법(Gradient Descent)
- [참고1-경사하강법](http://www.incodom.kr/%EA%B2%BD%EC%82%AC%ED%95%98%EA%B0%95%EB%B2%95), [참고2-경사하강법](https://yganalyst.github.io/ml/ML_chap3-2/#1-%EB%B0%B0%EC%B9%98-%EA%B2%BD%EC%82%AC%ED%95%98%EA%B0%95%EB%B2%95)
- 경사하강법
    - 학습률과 손실함수의 순간기울기(gradient)를이용하여 가중치(weight)를 업데이트하는 방법.
    - 미분의 기울기를 이용하여 도표의 오차들을 비교하고 오차를 최소화하는 방향으로 이동시키는 방법.
- 경사하강법 목적
    - 비용함수(평균제곱오차, mse)값이 최소가 되는 w와 b를 찾는 것
- 경사하강법 종류
    - 배치 경사하강법, 확률적 경사하강법(SGD), 미니배치 경사하강법

## 배치 경사하강법

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = 2*np.random.rand(10,1)
y = 4 + 3 * X + np.random.randn(10,1)
X_b = np.c_[np.ones((10,1)), X]    # 절편항을 위해 1열에 1값 추가

learning_rate = 0.1   # 학습률 설정
n_iter = 10          # 반복 횟수 설정
m = 100                 # 샘플 수

theta = np.random.randn(2,1)  # 무작위 초기값 설정

for iter in range(n_iter):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)  # 배치 경사하강법
    # print(gradients)
    theta = theta - learning_rate*gradients          # next step
    y_hat = X_b.dot(theta) # theta: weight

    plt.ylim([-3,5])
    plt.plot(X,y,'ro')
    plt.plot(X, y_hat) # n_iter를 늘려서 오차를 0으로 만들 때까지 실행해야 함.
    plt.show()
```
